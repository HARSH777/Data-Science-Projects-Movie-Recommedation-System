```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits = credits.rename(index = str, columns={'movie_id': 'id'})
```

```python
credits.head()
```

```python
mergedf = credits.merge(movies, on='id')
```

```python
mergedf.head()
mergeddfCleaned = mergeddf.drop(columns=['crew','genres','cast','homepage','keywords','overview','production_companies','production_countries','release_date','spoken_languages'], axis = 1)
```

```python
mergeddfCleaned.head()
```

```python
v=mergeddfCleaned['vote_count']
r=mergeddfCleaned['vote_average']
c=mergeddfCleaned['vote_average'].mean()
m=mergeddfCleaned['vote_count'].quantile(0.70)
mergeddfCleaned['weighted_average']= (r*v + c*m)/(v+m)
```

```python
plt.figure(figsize=[5,5])
sns.barplot(y = mergeddfCleaned['title_x'].head(), x = mergeddfCleaned['weighted_average'].head())
plt.xlabel = 'title'
plt.ylabel ='weightedAverage'
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scale = MinMaxScaler()
```

```python
mergeddfCleaned[['scaled_weighted_average','scaled_popularity']] = pd.DataFrame(scale.fit_transform(mergeddfCleaned[['weighted_average','popularity']]), columns=['weighted_average', 'popularity'])
```

```python
mergeddfCleaned['score'] = mergeddfCleaned['scaled_weighted_average']*0.5 +mergeddfCleaned['scaled_popularity']*0.5
```
