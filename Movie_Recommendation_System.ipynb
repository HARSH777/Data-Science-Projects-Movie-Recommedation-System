```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('u.data', sep='\t',names=['user_id', 'item_id','rating','timestamp'])
```

```python
dataset
```

```python
movie_title = pd.read_csv('Movie_Id_Titles')
```

```python
dataset = pd.merge(dataset,movie_title,on='item_id')
```

```python
dataset
```

```python
rating = pd.DataFrame(dataset.groupby('title')['rating'].mean())
```

```python
rating['number of rating'] = pd.DataFrame(dataset.groupby('title')['rating'].count())
```

```python
rating.sort_values('number of rating', ascending=False)
```

```python
movie_mat = pd.pivot_table(dataset, index = 'user_id', values='rating', columns='title')
```

```python
movie_mat
```

```python
similar_to_StarWars = movie_mat['Star Wars (1977)']
```

```python
similar_to_StarWars = movie_mat.corrwith(similar_to_StarWars)
```

```python
correlated = pd.DataFrame(similar_to_StarWars, columns=['Correlation']).sort_values('Correlation', ascending=False)
```

```python
final = pd.merge(correlated,rating,on= 'title')
```

```python
final[final['number of rating']>100].sort_values('Correlation', ascending=False)
```
